```python
import pandas as pd
```

```python
###importando el archivo
data = pd.read_csv('credit_card.csv')
data.head(1)
```

```python
###traduciendo las variables
titulos = {
    'LIMIT_BAL' : 'Limite',
    'CHECKING_ACCOUNT':'Cuenta_corriente',
    'EDUCATION':'Escolaridad',
    'MARRIAGE':'Estado_civil',
    'AGE':'Edad',
    'BILL_AMT':'Valor_de_factura',
    'PAY_AMT':'Valor_de_pago',
    'DEFAULT':'Moroso'
}
tarjetas = data.rename(columns = titulos)
tarjetas.head(1)
```

```python
#traduciendo datos
EspCuentaCorriente = {
    'Yes':'Si',
    'No':'No'
}
EspEscolaridad = {
    '2.University':'Universidad',
    '3.Graduate School': 'Posgrado', 
    '1.High School':'Colegio'
}
EspEstadoCivil= {
    'Married': 'Casado', 
    'Single': 'Soltero' 
}
cualitativa={
    1 : 'Si',
    0 : 'No'
}
tarjetas.Cuenta_corriente = tarjetas.Cuenta_corriente.map(EspCuentaCorriente)
tarjetas.Escolaridad = tarjetas.Escolaridad.map(EspEscolaridad)
tarjetas.Estado_civil = tarjetas.Estado_civil.map(EspEstadoCivil)
tarjetas.Moroso = tarjetas.Moroso.map(cualitativa)
tarjetas.head(1)
```

```python
#calculando una nueva variable
tarjetas['Indice_de_uso'] = tarjetas['Valor_de_factura'] / tarjetas['Limite']
tarjetas.head(5)
```

```python
#comparar los cambios
data.head(5)
```

```python
#visualizar estadistica descriptiva cantidad, promedio, desviacion, cuartiles y maximo 
tarjetas.Edad.describe()
```

```python
#extraer los datos del dataframe describe()
tarjetas.Edad.describe()['mean']
```

```python
#agrupar datos 
#sabiendo que los datos de la edad estan entre 21 y 75, vistos en el describe
grupos = [20,30,40,50,80]
rangos = ['entre 20 - 30', 'entre 30 - 40', 'entre 40 - 50', 'mayores a 50']
tarjetas['Rango_edad'] = pd.cut(tarjetas['Edad'], grupos, labels=rangos)
tarjetas
```

```python
#buscando registros apartir de un dato 
buscado = tarjetas.loc[tarjetas['Edad'] == 70]
buscado
```

```python
#exportar el dataframe como csv
tarjetas.to_csv('tarjetas.csv', sep=';',index=False)
```
